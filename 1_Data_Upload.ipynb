```python
import csv, operator
import os, shutil
from os import walk, getcwd

# Retorna todos los archivos de un directorio dado
def ls(ruta):  
    return next(walk(ruta))[2]

# Retorna la ruta destino según el tipo de celula
def ruta_tipo_cel(dataset_dir, idtipocelulasbase,identidadnombre):
    return {
        'CP': os.path.join(dataset_dir, 'ATYPICAL_LYMPHOCYTE', idtipocelulasbase),
        'ATYPICAL_LYMPHOCYTE': os.path.join(dataset_dir, 'ATYPICAL_LYMPHOCYTE', idtipocelulasbase),
        'PROLYMPHOCYTE': os.path.join(dataset_dir, 'ATYPICAL_LYMPHOCYTE', idtipocelulasbase)
    }.get(idtipocelulasbase,os.path.join(dataset_dir, idtipocelulasbase,))   

# Generamos estructura de directorios destino
dataset_dir = 'img'
original_dataset_dir = 'original_dataset'

atypical_lymphocyte_dir = os.path.join(dataset_dir,'ATYPICAL_LYMPHOCYTE')
os.mkdir(atypical_lymphocyte_dir)
aux_dir = os.path.join(atypical_lymphocyte_dir,'ATYPICAL_LYMPHOCYTE')
os.mkdir(aux_dir)
aux_dir =os.path.join(atypical_lymphocyte_dir,'CP')
os.mkdir(aux_dir)
aux_dir = os.path.join(atypical_lymphocyte_dir,'PROLYMPHOCYTE')
os.mkdir(aux_dir)


variant_lymphocyte_dir = os.path.join(dataset_dir,'VARIANT_LYMPHOCYTE')
os.mkdir(variant_lymphocyte_dir)

blast_dir  = os.path.join(dataset_dir,'BLAST')
os.mkdir(blast_dir)

lymphocyte_dir = os.path.join(dataset_dir, 'LYMPHOCYTE')
os.mkdir(lymphocyte_dir)
#aux_dir = os.path.join(lymphocyte_dir,'NORMAL')
#os.mkdir(aux_dir)

# Leer el archivo 'datos.csv' con reader() y 
# mostrar todos los registros, uno a uno:
with open('infos.csv') as csvarchivo:
    entrada = csv.reader(csvarchivo) # Abrir archivo csv
    for reg in entrada: # Leer todos los registros
        # Origen de datos
        identidadnombre, idtipocelulasbase, fecha, historia, nueva_ruta = next(entrada)
        src_dir = os.path.join(original_dataset_dir, identidadnombre, historia, fecha, idtipocelulasbase)
        print('Origen: ', src_dir)
        orig_archives = (ls(ruta = src_dir))
        #print(orig_archives)
        
        #Destino de datos
        dest_dir = (ruta_tipo_cel(dataset_dir, idtipocelulasbase,identidadnombre))
        #dest_dir = os.path.join(dataset_dir, idtipocelulasbase)
        print('Destino: ', dest_dir)
       
        for archivo in orig_archives:
            orig = os.path.join(src_dir, archivo)
            #print('Origen con archivo: ', orig)
            dest = os.path.join(dest_dir, archivo)
            #print('Destino con archivo: ', dest)
            shutil.copyfile(orig, dest)
            
del  identidadnombre, idtipocelulasbase, fecha, historia, aux_dir, nueva_ruta, entrada, orig, dest
del src_dir, dest_dir, archivo
csvarchivo.close()  # Cerrar archivo
del csvarchivo  # Borrar objeto





```
